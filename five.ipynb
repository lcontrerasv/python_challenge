```python
banner = 'http://www.pythonchallenge.com/pc/def/banner.p'
```

```python
urlretrieve(banner, filename = 'level5.pkl')
```

```python
with open('level5.pkl', 'rb') as fp:
    result = pickle.load(fp)
    print(result)
```

```python
for i in result:
    for j in i:
        print(j[0] * j[1], end = '')
    print('\n')
```
